# CIFRADO DE SUSTITUCIÓN(CIFRADO CÉSAR)

* Se basa en sustituir cada letra del mensaje por la letra que se encuentra un número fijo de posiciones más adelante en el alfabeto. Este número fijo se llama la clave de desplazamiento.



# FUNCIÓN

```python
import string

# Definimos el alfabeto base para obtener el índice de cada letra (0-25)
ALFABETO_BASE = string.ascii_uppercase # 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
LONGITUD_ALFABETO = len(ALFABETO_BASE) # Es 26

def cifrado_cesar(texto, clave, modo='cifrar'):
    """
    Cifra o descifra un texto usando el Cifrado César.

    Parámetros:
    - texto (str): El mensaje a procesar.
    - clave (int): El número de posiciones a desplazar (ej., 3).
    - modo (str): 'cifrar' para desplazar hacia adelante, 'descifrar' para atrás.
    """

    resultado = ""

    # Si estamos descifrando, el desplazamiento debe ser en sentido contrario
    if modo == 'descifrar':
        clave = -clave

    print(f"Paso 1: La clave de desplazamiento efectiva es: {clave}")

    # Procesamos el texto, convirtiendo a mayúsculas
    for char in texto.upper():

        if char in ALFABETO_BASE:
            # 1. Encontrar la posición de la letra en el alfabeto (0-25)
            indice_original = ALFABETO_BASE.index(char)

            # 2. Aplicar el desplazamiento y el "módulo" (el envoltorio)
            # El % 26 asegura que volvemos a la 'A' después de la 'Z'
            indice_cifrado = (indice_original + clave) % LONGITUD_ALFABETO

            # 3. Obtener la nueva letra y añadirla al resultado
            letra_cifrada = ALFABETO_BASE[indice_cifrado]
            resultado += letra_cifrada

        else:
            # Si no es una letra (espacio, !, ?, etc.), la conservamos
            resultado += char

    return resultado
```

# PRUEBA

```python
# --- Ejemplo para clase ---
mensaje = "UNIVERSIDAD CENTRAL DEL ECUADOR"
clave = 5

print(f"Mensaje de Entrada: {mensaje}")
print(f"Clave de Desplazamiento: {clave}\n")

print("--- PROCESO DE CIFRADO (Desplazamiento +5) ---")
texto_cifrado = cifrado_cesar(mensaje, clave, modo='cifrar')
print(f"\nTEXTO CIFRADO: {texto_cifrado}")

print("\n" + "=" * 30 + "\n")

print("---PROCESO DE DESCIFRADO (Desplazamiento -5) ---")
# Usamos la misma clave, pero la función internamente la convierte a -5
texto_descifrado = cifrado_cesar(texto_cifrado, clave, modo='descifrar')
print(f"\nTEXTO DESCIFRADO: {texto_descifrado}")
```

# FUNCIONAMIENTO

1. Mapeo de Posiciones
Imagina que cada letra tiene un número (su índice):

```python
# A,B,C,D,E,...,V,W,X,Y,Z
# 0,1,2,3,4,...,21,22,23,24,25
```

2. La Operación
Para cifrar una letra, aplicamos la fórmula:

Ejemplo con 'Z' (Índice 25) y Clave 5 (El Envoltorio):$$ \\ (25 + 5) \pmod{26} = 30 \pmod{26} = 4
$$$$$$El índice 4 es la E. (Z $\rightarrow$ E) El módulo 26 nos lleva de la Z a la E automáticamente.

* El proceso para descifrar el Cifrado César funciona de manera inversa al cifrado. En lugar de sumar la clave de desplazamiento, se resta (o se desplaza hacia atrás).
